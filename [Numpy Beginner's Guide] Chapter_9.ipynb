```python
import matplotlib.pyplot as plt
```

```python
func = np.poly1d(np.array([1, 2, 3]).astype(float))
```

```python
x = np.linspace(-10, 10, 30)
```

```python
y = func(x)
```

```python
pyplot.plot(x, y)pyplot.xlabel('x')plt.ylabel('y(x)')
```

```python
func1 = func.deriv(m=1)
```

```python
y = func(x)
```

```python
y1 = func1(x)
```

```python
plt.plot(x, y, 'ro', x, y1, 'g--')plt.xlabel('x')plt.ylabel('y')
```

```python
plt.subplot(3, 1, 1)plt.plot(x, y, 'r-')plt.title('Poly')plt.subplot(3, 1, 2)plt.plot(x, y1, 'b^')plt.title('First_Derv')plt.subplot(3, 1, 3)func2 = func.deriv(m=2)y2 = func2(x)plt.plot(x, y2, 'go')plt.title('Second_Derv')
```
